```python
def read_signals():

    with open(join(cf.DATA_DIR,'signal.dat'),'r') as d:
        signal = d.readlines()

    signal = np.array(signal)
    
    plt.plot(signal[0:100],'r-')
    plt.title('Raw Signal')
    plt.show()
    
    df = pd.DataFrame(signal[0:100])
    
    smooth_data = df.rolling(window=5).mean().plot(title='Moving Average', style='k')
    plt.show()
    return None

read_signals()
```
